```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:\\Users\\ujwal\\Downloads\\5g-Data-Predictor\\5G_Data_Usage_Prediction_Enhanced.csv"
df = pd.read_csv(file_path)

# --- Basic Overview ---
print("Head of Dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nDescriptive Statistics for Numeric Columns:")
print(df.describe())

# --- Categorical Analysis ---
# List of categorical columns
categorical_columns = [
    'device_type', 'app_category', 'data_quality',
    'time_of_day', 'day_of_week', 'network_type',
    'location_type', 'indoor_outdoor'
]

# Plot count plots for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# --- Numeric Distribution ---
# List of numeric columns
numeric_columns = [
    'session_duration', 'signal_strength', 'prev_usage',
    'throughput', 'background_data', 'data_usage'
]

# Plot histograms for each numeric variable
df[numeric_columns].hist(bins=15, figsize=(15, 10), layout=(2, 3))
plt.suptitle("Histograms of Numeric Columns")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


# --- Additional Analysis Example: Pie Chart for a Categorical Column ---
def plot_pie_chart(column_name):
    data_counts = df[column_name].value_counts()
    labels = data_counts.index
    sizes = data_counts.values
    # Explode the slice with the maximum value to highlight it
    explode = [0.1 if size == sizes.max() else 0 for size in sizes]
    
    plt.figure(figsize=(9, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140,
            colors=plt.cm.Paired.colors, explode=explode, shadow=True)
    plt.title(f"Distribution of {column_name}", fontsize=14, fontweight='bold')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.show()

# Example: Pie chart for the 'device_type' column
plot_pie_chart('device_type')
```
